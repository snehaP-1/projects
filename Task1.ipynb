```python
import pandas as pd
import numpy as np
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.metrics import classification_report
from transformers import pipeline

# **Download necessary NLTK data**
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# **Step 1: Load Dataset**
def load_data(file_path):
    """Loads dataset from a CSV file and fixes missing headers."""
    df = pd.read_csv(file_path, encoding="ISO-8859-1", header=None)
    
    # Manually set correct column names
    df.columns = ["Sentiment", "Sentence"]
    
    print("Loaded dataset successfully with columns:", df.columns.tolist())  
    return df

# **Step 2: Preprocess Text**
def preprocess_text(text):
    """Cleans text by removing special characters, stopwords, and lemmatizing."""
    if not isinstance(text, str):
        return ""
    
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Keep only letters and spaces
    tokens = word_tokenize(text.lower())  # Tokenize
    stop_words = set(stopwords.words('english'))  # Load stopwords
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Lemmatize
    return " ".join(tokens)

# **Step 3: Sentiment Analysis using BERT**
def predict_sentiment(texts):
    """Uses BERT-based model to analyze sentiment."""
    sentiment_model = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
    return sentiment_model(texts)

# **Step 4: Evaluate Model Performance**
def evaluate_model(predictions, true_labels):
    """Evaluates the sentiment model using classification metrics."""
    label_mapping = {'positive': 1, 'negative': 0, 'neutral': 2}  
    true_labels_num = [label_mapping.get(label.lower(), 2) for label in true_labels] 

    predicted_labels = []
    for pred in predictions:
        if pred['label'] == 'POSITIVE':
            predicted_labels.append(1)
        elif pred['label'] == 'NEGATIVE':
            predicted_labels.append(0)
        else:
            predicted_labels.append(2)

    print("\nClassification Report:\n", classification_report(true_labels_num, predicted_labels))

# **Main Execution**
if __name__ == "__main__":
    file_path = r"C:\Users\om\Desktop\Internship\Dataset\all-data.csv"  # Update the path if necessary
    
    # Load data
    df = load_data(file_path)

    # Extract and Verify Columns
    news_articles = df["Sentence"].astype(str).tolist()
    labels = df["Sentiment"].astype(str).tolist()

    # Preprocess Text Data
    preprocessed_articles = [preprocess_text(article) for article in news_articles]

    # Predict Sentiments
    predictions = predict_sentiment(preprocessed_articles)

    # Evaluate Model
    evaluate_model(predictions, labels)

    # Display Sample Predictions
    print("\nSample Predictions:")
    for article, prediction in zip(news_articles[:5], predictions[:5]):
        print(f"Text: {article}\nPrediction: {prediction}\n")
```

```python
import pandas as pd
import numpy as np
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.metrics import classification_report
from transformers import pipeline

# Download necessary NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# Step 1: Load Data
def load_data(file_path):
    try:
        df = pd.read_csv(file_path, encoding="utf-8", errors="replace")
        print("Dataset Loaded Successfully!")
        print(df.head())  # Display first few rows
        return df
    except Exception as e:
        print(f"Error loading dataset: {e}")
        return None

# Step 2: Preprocess Text Data
def preprocess_text(text):
    if isinstance(text, str):
        text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove special characters and numbers
        tokens = word_tokenize(text.lower())  # Tokenize and convert to lowercase
        tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Lemmatization
        return " ".join(tokens)
    return ""

# Step 3: Sentiment Analysis using BERT (Transformers)
def predict_sentiment(texts):
    sentiment_model = pipeline("sentiment-analysis", model="distilbert-base-uncased")
    sentiments = sentiment_model(texts)
    return sentiments

# Step 4: Evaluate Model
def evaluate_model(predictions, true_labels):
    try:
        true_labels_num = [1 if label.lower() == 'positive' else 0 for label in true_labels]
        predicted_labels = [1 if pred['label'] == 'POSITIVE' else 0 for pred in predictions]
        print(classification_report(true_labels_num, predicted_labels))
    except Exception as e:
        print(f"Error in evaluation: {e}")

# Main Execution
if __name__ == "__main__":
    file_path = "/mnt/data/all-data.csv"  # Ensure correct file path
    df = load_data(file_path)

    if df is not None:
        # Print column names to check structure
        print("Column Names:", df.columns)

        # Ensure required columns exist
        if 'Sentence' in df.columns and 'Sentiment' in df.columns:
            news_articles = df['Sentence'].astype(str).tolist()  # Convert to string in case of NaNs
            labels = df['Sentiment'].astype(str).tolist()  # Ensure labels are strings

            # Step 2: Preprocess the text data
            preprocessed_articles = [preprocess_text(article) for article in news_articles]

            # Step 3: Predict sentiment
            predictions = predict_sentiment(preprocessed_articles)

            # Step 4: Evaluate the model
            evaluate_model(predictions, labels)

            # Display sample results
            for article, prediction in zip(news_articles[:5], predictions[:5]):
                print(f"Article: {article}\nPrediction: {prediction}\n")
        else:
            print("Error: Required columns ('Sentence' and 'Sentiment') not found in the dataset.")
```
